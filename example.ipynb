# example for using learning rate decay schedualr in lr_schedular.py

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

import model
```

```python
# load model
model = model.model(8, 12)
```

## 1. CosineAnnealingCyclicalLR

```python
# load optimizor
optimizer = torch.optim.Adam(
            params=[{'params': model.parameters()}],
            lr=1e-4
        )

# init learning rate schedular
import lr_scheduler
lr = lr_scheduler.CosineAnnealingCyclicalLR(optimizer,T_max=50,decay=0.666,T_multi=1,eta_min=1e-6)

# step
lr_epoch = np.zeros([200])
for epoch in range(200):
    lr_epoch[epoch] = optimizer.param_groups[0]['lr']
    lr.step()

plt.plot(range(200),lr_epoch)
plt.grid(True)
plt.show()
```

## 2. TrangleCyclicalLR

```python
# load optimizor
optimizer = torch.optim.Adam(
            params=[{'params': model.parameters()}],
            lr=1e-4
        )

# init learning rate schedular
import lr_scheduler
lr = lr_scheduler.TrangleCyclicalLR(optimizer,T_max=50,eta_min=1e-5)

# step
lr_epoch = np.zeros([200])
for epoch in range(200):
    lr_epoch[epoch] = optimizer.param_groups[0]['lr']
    lr.step()

plt.plot(range(200),lr_epoch)
plt.grid(True)
plt.show()
```

## 3. TrangleTailLR

```python
# load optimizor
optimizer = torch.optim.Adam(
            params=[{'params': model.parameters()}],
            lr=1e-4
        )

# init learning rate schedular
import lr_scheduler
lr = lr_scheduler.TrangleTailLR(optimizer,T_max=200,eta_min=1e-5)

# step
lr_epoch = np.zeros([200])
for epoch in range(200):
    lr_epoch[epoch] = optimizer.param_groups[0]['lr']
    lr.step()

plt.plot(range(200),lr_epoch)
plt.grid(True)
plt.show()
```
